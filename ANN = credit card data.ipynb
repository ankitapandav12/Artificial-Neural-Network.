```python
X=credit.iloc[:,1:29].values
X
```

```python
y=credit.iloc[:,-1].values
y
```

```python
from imblearn.over_sampling import SMOTE
ros = SMOTE()
```

```python
x_data,y_data=ros.fit_resample(X,y)
```

```python
from collections import Counter
print("Original Sample {}".format(Counter(y)))
print("Resampling Data {}".format(Counter(y_data)))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=10, activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=10, activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=10, activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
ann.compile(optimizer = 'adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=1000,epochs=100)
```

```python
y_pred=ann.predict(x_test)
y_pred =(y_pred > 0.5)

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
ac=accuracy_score(y_test, y_pred)*100
print(ac)
```
